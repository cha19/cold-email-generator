```python
from langchain_groq import ChatGroq
from dotenv import load_dotenv
import os
from langchain_core.output_parsers import JsonOutputParser

import pandas as pd
import chromadb
import uuid
```

```python
load_dotenv()
api_key=os.getenv('groq_api_key')
```

```python
llm = ChatGroq(
    model_name = 'llama-3.3-70b-versatile',
    temperature =0,
    groq_api_key = api_key
)

response = llm.invoke("the first person to land on the moon")
print(response.content)
```

## For Webscarpping    

```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://careers.nike.com/principal-software-engineer/job/R-45778")
page_data = loader.load().pop().page_content
print(page_data, sep='\n\n')

```

```python
print(page_data)
```

```python
from langchain_core.prompts import PromptTemplate
```

```python
prompt_extract = PromptTemplate.from_template(
    """
    ### SCRAPED TEXT FROM WEBSITE:
    {page_data}
    ### INSTRUCTION:
    the scraped text is from the career's page of a website.
    Your job is to extract the job postings and return them in JSON format containing following keys: 'role','experience','skills' and 'description'.
    Only return the valid JSON.
    ### VALID JSON(NO PREAMBLE):
    """
)
```

```python
chain_extract = prompt_extract | llm
res = chain_extract.invoke(input={'page_data':page_data})
print(res.content)
```

```python
json_parser = JsonOutputParser()
json_res = json_parser.parse(res.content)
json_res
```

```python
df = pd.read_csv('./app/resources/my_portfolio.csv')
```

```python
df
```

```python
client = chromadb.PersistentClient('vectorstore')
collection = client.get_or_create_collection(name="portfolio")

if not collection.count():
    for _, row in df.iterrows():
        collection.add(documents=row["Techstack"],
                       metadatas={"links": row["Links"]},
                       ids=[str(uuid.uuid4())])
```

```python
links = collection.query(query_texts=['Experience in Python','Expertise in React Native'], n_results=2).get('metadatas', [])
links
```

```python
links = collection.query(query_texts=job['skills'], n_results=2).get('metadatas', [])
links
```

```python
prompt_email = PromptTemplate.from_template(
        """
        ### JOB DESCRIPTION:
        {job_description}
        
        ### INSTRUCTION:
        You are charan, a business development executive at UdbhavX. UdbhavX is an AI & Software Consulting company dedicated to facilitating
        the seamless integration of business processes through automated tools. 
        Over our experience, we have empowered numerous enterprises with tailored solutions, fostering scalability, 
        process optimization, cost reduction, and heightened overall efficiency. 
        Your job is to write a cold email to the client regarding the job mentioned above describing the capability of AtliQ 
        in fulfilling their needs.
        Also add the most relevant ones from the following links to showcase UdbhavX's portfolio: {link_list}
        Remember you are charan, SDE at UdbhavX. 
        Do not provide a preamble.
        ### EMAIL (NO PREAMBLE):
        
        """
        )

chain_email = prompt_email | llm
res = chain_email.invoke({"job_description": str(job), "link_list": links})
print(res.content)
```
